```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

**Daily Dataset**

```python
df1= pd.read_csv("/content/daily_merged_cleaned.csv")
```

```python
df1
```

```python
top_10_df1 = df1.sort_values(by='TotalSteps', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(top_10_df1['ActivityDate'], top_10_df1['TotalSteps'])
plt.title('TotalSteps per Date')
plt.xlabel('ActivityDate')
plt.ylabel('TotalSteps')
plt.show()
```

```python
# Grouping by Date and summing up TotalSteps and TotalDistance
grouped_df1 = df1.groupby('ActivityDate').agg({'TotalSteps': 'sum', 'TotalDistance': 'sum'}).reset_index()
# Plotting the Line Chart
plt.figure(figsize=(12, 6))
plt.plot(grouped_df1['ActivityDate'], grouped_df1['TotalSteps'], label='Total Steps', marker='o')
plt.plot(grouped_df1['ActivityDate'], grouped_df1['TotalDistance'], label='Total Distance', marker='o')
plt.xlabel('ActivityDate')
plt.ylabel('Count')
plt.title('Total Steps vs Total Distance over Time')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Sorting the DataFrame by TotalSteps in descending order and selecting the top 5 rows
top_5_df1 = df1.sort_values(by='TotalSteps', ascending=False).head(5)

# Creating subplots for each metric
fig, axes = plt.subplots(3, 3, figsize=(15, 12))

# Bar Chart for TotalSteps
axes[0, 0].bar(top_5_df1['ActivityDate'], top_5_df1['TotalSteps'])
axes[0, 0].set_title('TotalSteps')

# Bar Chart for VeryActiveMinutes
axes[0, 1].bar(top_5_df1['ActivityDate'], top_5_df1['VeryActiveMinutes'])
axes[0, 1].set_title('VeryActiveMinutes')

# Bar Chart for TotalDistance
axes[0, 2].bar(top_5_df1['ActivityDate'], top_5_df1['TotalDistance'])
axes[0, 2].set_title('TotalDistance')

# Bar Chart for SedentaryMinutes
axes[1, 0].bar(top_5_df1['ActivityDate'], top_5_df1['SedentaryMinutes'])
axes[1, 0].set_title('SedentaryMinutes')

# Bar Chart for FairlyActiveMinutes
axes[1, 1].bar(top_5_df1['ActivityDate'], top_5_df1['FairlyActiveMinutes'])
axes[1, 1].set_title('FairlyActiveMinutes')

# Bar Chart for LightlyActiveMinutes
axes[1, 2].bar(top_5_df1['ActivityDate'], top_5_df1['LightlyActiveMinutes'])
axes[1, 2].set_title('LightlyActiveMinutes')

# Bar Chart for VeryActiveDistance
axes[2, 0].bar(top_5_df1['ActivityDate'], top_5_df1['VeryActiveDistance'])
axes[2, 0].set_title('VeryActiveDistance')

# Bar Chart for ModeratelyActiveDistance
axes[2, 1].bar(top_5_df1['ActivityDate'], top_5_df1['ModeratelyActiveDistance'])
axes[2, 1].set_title('ModeratelyActiveDistance')

# Bar Chart for LightActiveDistance
axes[2, 2].bar(top_5_df1['ActivityDate'], top_5_df1['LightActiveDistance'])
axes[2, 2].set_title('LightActiveDistance')

# Adjusting layout for better readability
plt.tight_layout()
plt.show()
```

**Minute Dataset**

```python
df4= pd.read_csv("/content/minute_merged_cleaned.csv")
```

```python
df4
```

```python
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Id']
  ys = series['METs']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(5, 5), layout='constrained')
df_sorted = df4.sort_values('Id', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Id')
_ = plt.ylabel('METs')
```

```python
# Selecting relevant columns for the pair plot
columns_for_pair_plot = ['Steps', 'Calories', 'Intensity']

# Pair Plot
sns.pairplot(df4[columns_for_pair_plot])
plt.suptitle('Pair Plot: Steps, Calories, and Intensity', y=1.02)
plt.show()
```
